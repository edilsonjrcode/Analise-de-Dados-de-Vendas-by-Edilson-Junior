## Importando as bibliotecas

```python
import pandas as pd
import numpy as np
```

## Criando uma tabela de vendas

```python
venda = {'Loja': ['Tambiá Shopping','Manaíra Shopping','Mangabeira Shopping','Mag Shopping']*4,
         'Produto': ['Iphone','iPod 256GB','Macbook Pro','Xiaomi Redmi Note 8',
                     'iPod 256GB','Macbook Pro','Xiaomi Redmi Note 8','Iphone',
                     'Macbook Pro','Xiaomi Redmi Note 8','Iphone','iPod 256GB',
                     'Xiaomi Redmi Note 8','Iphone','iPod 256GB','Macbook Pro'], 
         'Código do Produto': ['44125','55876','423','4421','44125','55876','423','4421']*2,
         'Quantidade' : np.random.randint(1,5, size=16), 
         'Valor do Produto': [3500,950,12500,1400,950,12500,1400,3500,12500,1400,3500,950,1400,3500,950,12500]}

venda_df = pd.DataFrame(venda)
display(venda_df)

venda_df.to_csv('Vendas.csv', index =False) #armazenando os dados em um arquivo Vendas.csv
```

## Criando uma nova coluna

```python
venda_df['Faturamento'] = venda_df['Quantidade'] * venda_df['Valor do Produto']
display(venda_df)
```

## Produto mais vendido 

```python
produto_mais_vendido = venda_df[['Produto', 'Quantidade']].groupby(['Produto']).sum()
display(produto_mais_vendido)
```

## Faturamento Total

```python
faturamento = venda_df['Faturamento'].sum()
print(f'O faturamento da empresa foi de R${faturamento}')
```

## Faturamento por Loja

```python
faturamento_por_loja = venda_df[['Loja', 'Faturamento']].groupby(['Loja']).sum()
display(faturamento_por_loja)
```

## Fauramento por Produto

```python
faturamento_por_produto = venda_df[['Produto', 'Faturamento']].groupby(['Produto']).sum()
display(faturamento_por_produto)
```

## Faturamento da Loja por Produto

```python
faturamento_loja_produto = venda_df[['Loja','Produto', 'Faturamento']].groupby(['Loja','Produto']).sum()
display(faturamento_loja_produto)
```
